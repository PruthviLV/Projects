# Lab test link 1 scrapping

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
url='https://www.martinhealth.org/common-lab-tests-mhs'
r=requests.get(url)
print(r.content)
```

```python
soup=BeautifulSoup(r.content, 'html5lib')
print(soup.prettify())
```

```python
meta = soup.find('meta', {"name": "description"}).attrs['content'].strip("Common Lab Tests - Martin Health System")
values=meta.replace(":","")
final_values=values.split(",")
list1=[]
for l in final_values:
    list1.append(l.replace("\n",""))
df = pd.DataFrame({'Class_Name':list1})
dfnew1=df.assign(Parent_class="Lab_tests", Relation="Subclass")
dfnew1['Class_Name'] = dfnew1['Class_Name'].str.strip()
dfnew1
```

# Lab test link 2 scrapping

```python
url='https://www.medifee.com/dc/7670-laboratory-diagnosis/'
r=requests.get(url)
#print(r.content)
soup=BeautifulSoup(r.content, 'lxml')
soup
```

```python
#The following line will generate a list of HTML content for each table
table = soup.find_all("table", attrs={"class": "table table-bordered"})
print(table)
```

```python
#scrapping the fist column
table1=table[0]
# the head will form our column names
body=table1.find_all("tr")
head=body[0]# 0th item is the header row
body_rows = body[1:] # All other items becomes the rest of the rows
```

```python
headings = []
for item in head.find_all("th"): # loop through all th elements
    # convert the th elements to text and strip "\n"
    item = (item.text)
    # append the clean column name to headings
    headings.append(item)
print(headings)
```

```python
all_rows = [] # will be a list for list for all rows
for row_num in range(len(body_rows)): # A row at a time
    row = [] # this will old entries for one row
    for row_item in body_rows[row_num].find_all("td"): #loop through all row entries
        # row_item.text removes the tags from the entries
        # the following regex is to remove \xa0 and \n and comma from row_item.text
        # xa0 encodes the flag, \n is the newline and comma separates thousands in numbers
        aa = row_item.text
        #append aa to row - note one row entry is being appended
        row.append(aa)
    # append one row to all_rows
    all_rows.append(row)
print(all_rows)
```

```python
df = pd.DataFrame(data=all_rows,columns=headings)
dataframe = df.drop(['View Price'], axis=1)
dataframe.rename(columns = {'Diagnostic Test Name':'Class_Name'}, inplace = True)
dfnew2=dataframe.assign(Parent_class="Lab_tests", Relation="Subclass")
dfnew2['Class_Name'] = dfnew2['Class_Name'].str.strip()
dfnew2
```

# Merging both the link classes and dropping duplicates

```python
result=pd.concat([dfnew1,dfnew2])
result=result.reset_index(drop=True)
res_1=result.drop_duplicates(keep='last')
data1=res_1.drop(labels=[8,13,14], axis=0)
data1=data1.reset_index(drop=True)
data1
```
