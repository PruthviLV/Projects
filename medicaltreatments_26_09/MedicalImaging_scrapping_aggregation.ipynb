# Medical Imaging Link 1 scrapping

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
url='https://www.britannica.com/topic/list-of-medical-tests-and-diagnostic-procedures-2074273#ref335439'
r=requests.get(url)
soup=BeautifulSoup(r.content, 'lxml')
soup
```

```python
list1=[]
parent = []
medicalImaging = []
for ele in soup.find(id = 'ref335439').find_next("ul").find_all("li"):
    #print(ele)
    if ele.find("ul"):
        #print(ele.find("ul"))
        if ele.text not in medicalImaging:
            subparent=ele.find("a").text
            medicalImaging.append(subparent)
            parent.append("Medical_imaging")
        for subelement in ele.find("ul").find_all("li"):
            if ele.text not in medicalImaging:
                medicalImaging.append(subelement.text)
                parent.append(subparent)
    else:
        if ele.text not in medicalImaging:
            medicalImaging.append(ele.text)
            parent.append("Medical_imaging")
dfnew1 = pd.DataFrame({"Class_Name":medicalImaging, "Parent_class": parent, "Relation": "subClass"})
dfnew1
```

# Medical Imaging Link 2 scrapping

```python
url='https://www.physio-pedia.com/Medical_Imaging'
r=requests.get(url)
soup=BeautifulSoup(r.content, 'lxml')
soup
```

```python
list1=[]
parent = []
medicalImaging = []
for ele in soup.find(id = 'Optoacoustic_Imaging').find_next("ul").find_all("li"):
    #print(ele)
    if ele.find("ul"):
        #print(ele.find("ul"))
        if ele.text not in medicalImaging:
            subparent=ele.text.split("\n")[0]
            medicalImaging.append(subparent)
            parent.append("Medical_imaging")
        for subelement in ele.find("ul").find_all("li"):
            if ele.text not in medicalImaging:
                medicalImaging.append(subelement.text)
                parent.append(subparent)
    else:
        if ele.text not in medicalImaging:
            medicalImaging.append(ele.text)
            parent.append("Medical_imaging")
dfnew2 = pd.DataFrame({"Class_Name":medicalImaging, "Parent_class": parent, "Relation": "subClass"})
dfnew2
```

# Merging both the link classes and dropping duplicates

```python
result = pd.concat([dfnew1,dfnew2])
result = result.reset_index(drop=True)
data2=result.drop_duplicates(keep='first')
data2
```
